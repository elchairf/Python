## Adding Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y,ls="-",marker='o',c='y')

plt.xlabel("delivery price")
plt.ylabel("weight")
plt.title("delivery prices by weight")

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])


font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x, y,ls="-",marker='o',c='y')

plt.xlabel("delivery price", fontdict = font2)
plt.ylabel("weight",fontdict = font2)
plt.title("delivery prices by weight", fontdict = font1, loc = 'left') # loc - location of the label

plt.show()

```

```python
plt.grid() #plt.grid(axis = 'x') #plt.grid(axis = 'y')
#plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.plot(x, y,ls="-",marker='o',c='y')

plt.xlabel("delivery price", fontdict = font2)
plt.ylabel("weight",fontdict = font2)
plt.title("delivery prices by weight", fontdict = font1, loc = 'left') # loc - location of the label

plt.show()
```

## multiple plots

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)  # creates a figure with 1 row (left argument), 2 columns (middle argument) 
                      #and the plot will be located in the first cell
                      # plt.subplot(2, 1, 1) --> 2 rows and 1 col
plt.title("left-side")
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) # the next diagram is plotted in the second cell
plt.plot(x,y)
plt.title("right-side")

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y, color = colors)

plt.show()
```

## color map

יש כמה סקלות צבע מובנות - לדוגמא
<br>viridis

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
fuelcons = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])


plt.scatter(x, y, c=colors, cmap='BuGn',s =fuelcons) ## cmap = 'Accent', 'Blues', 'BuGn', 'Oranges', 'Paired'
                                            ## to reverse the colors add '_r': 'Accent_r'
plt.colorbar() ## adds the color bar
plt.show()
```

### Marker sizes

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)
plt.show()

```

### transparent markers

```python
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

## combine color and sizes:

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

## Pie chart

```python
y = np.array([35, 25, 25, 15])

plt.pie(y,autopct='%1.0f%%')
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 180)
plt.show() 
```

```python
# pull slices away

myexplode = [0.2, 0, 0.4, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
```

```python
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors) # set colors

plt.legend() # add legend; 
#plt.legend(title = "Fruits:")

```
